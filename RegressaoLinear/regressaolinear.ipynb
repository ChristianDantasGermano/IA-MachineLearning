#Configurações

##Importações

```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

##Capturando tabela de dados 'House'

```python
 df = pd.read_csv('./house.csv',';') 
```

#Preparação dos dados

**Dados que serão utilizados**

```python
df
```

**Colunas dos dados**

```python
df.info()
```

**Definição das colunas de feature**

```python
feature_cols = ['houseSize','lotSize','bedrooms','granite','bathroom'] 
X_train = df[feature_cols] 
X_train.columns 
```

**Definição da coluna Target**

```python
y_train = df.sellingPrice
df.columns[5]  
```

**Separação dos dados de treinamento com proporção de 70% treinamento e 30% testes**

```python
X_train, X_test, y_train, y_test = train_test_split(X_train, y_train, test_size=0.3, random_state=1)
```

```python
X_train
```

```python
X_test
```

#Aplicação do Algoritmo

**Aplicação do modelo de regressão linear**

```python
model = LinearRegression()
```

**Treinamento do modelo**

```python
model.fit(X_train, y_train)
```

**Calculando o coeficiente de determinação R2, quanto mais perto de 1 melhor**

```python
r_sq = model.score(X_train, y_train)
print('coefficient of determination:', r_sq)
```

**Cálculo de MSE**

```python
np.mean((model.predict(X_train)-y_train)**2)
```

**Equação da Regressão Linear: 𝑓(𝑥) = 𝑏₀ + 𝑏₁𝑥**



```python
print('b0 ou intercept:', model.intercept_)
print('b1 ou coeficiente (slope):', model.coef_)
```

**Fazendo testa manual de predição, após o treinamento.**

```python
x_prev = np.array([[4000,16000,6,1,1]])
y_prev = model.predict(x_prev)
print("Valor de venda esperado da casa: R$ {0:2.2f}".format(y_prev[0]))
```
